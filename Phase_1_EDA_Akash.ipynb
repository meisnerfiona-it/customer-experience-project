```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Optional: improve plot appearance
sns.set_style("whitegrid")
```

```python
df_demo = pd.read_csv("/Users/akashkumarsamantray/Desktop/df_final_demo.txt")
web_1 = pd.read_csv("/Users/akashkumarsamantray/Desktop/df_final_web_data_pt_1.txt")
web_2 = pd.read_csv('/Users/akashkumarsamantray/Desktop/df_final_web_data_pt_2.txt')

print("demo shape:", df_demo.shape)
print("web_1 shape:", web_1.shape)
print("web_2 shape:", web_2.shape)
```

```python
##print(df_demo.head())
print(df_demo.columns)
```

```python
## EDA: Client Age Distribution
```

```python
# Median age
age_median = df_demo['clnt_age'].median()

# Plot age distribution
plt.figure(figsize=(10,6))

sns.histplot(df_demo['clnt_age'], bins=30, kde=True)

# Median line
plt.axvline(age_median, color='red', linestyle='--',
            label=f'Median Age: {age_median:.1f}')

# Labels and title
plt.title('Client Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')

plt.legend()

plt.show()

```

```python
### Insight:
1. Most clients appear to be middle-aged to older adults, with the median age concentrated around the central age range.
2. The median client age is 47 years, indicating that most clients are middle-aged rather than very young customers.
```

```python
## EDA: Client Tenure Distribution
```

```python
# Median tenure
tenure_median = df_demo['clnt_tenure_yr'].median()

# Plot tenure distribution
plt.figure(figsize=(10,6))

sns.histplot(df_demo['clnt_tenure_yr'], bins=30, kde=True)

# Median line
plt.axvline(tenure_median, color='red', linestyle='--',
            label=f'Median Tenure: {tenure_median:.1f} years')

# Labels and title
plt.title('Client Tenure Distribution')
plt.xlabel('Tenure (Years)')
plt.ylabel('Count')

plt.legend()

plt.show()

print(tenure_median)
```

```python
### Insight:
1. Most clients appear to be long-standing customers, indicating strong customer retention over time.
2. The median client tenure is 11 years, suggesting that many clients are long-standing customers with strong retention over time.    
```

```python
# Conclusion
```

```python
The analysis shows that the client base mainly consists of middle-aged individuals with long-standing relationships with the company. This may indicate a stable and loyal customer population.
```
